```python
from transformers import pipeline
classifier = pipeline('translation', model="facebook/wmt19-ru-en") # using the translartion model pipeline of the transformers library 
# to implement WMT19 russian to english translation NLP task
```

```python
russian_text_list = ['реализована 5-я модель', 'Спасибо за ответ']
classifier(russian_text_list)
```
